# Gesture recognition with random forest

### Libraries

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

### Load data

```python
df = pd.read_csv('datasets/1667171758.794016.csv').drop(['Unnamed: 0'], axis=1)
df.head(10)
```

```python
df.tail(10)
```

```python
x = df.drop(['Y'], axis=1)
y = df['Y']
print(x.shape, y.shape)
```

### Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.9, random_state=1, stratify=y)
```

### Model

```python
forest = RandomForestClassifier(criterion='log_loss',
n_estimators=50,
n_jobs=-1)
```

### Train model

```python
forest.fit(X_train, y_train)
```

### Predict test

```python
y_pred = forest.predict(X_test)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```
